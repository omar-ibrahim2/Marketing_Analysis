```python
-- Explore the data in the table
SELECT *
FROM 'marketing_data.csv'
```

```python
-- Who are the highest spending customers?
select
    ID,
    (NumWebPurchases + NumStorePurchases) as total_purchases,
    Income,
    (MntFishProducts + MntFruits + MntWines + MntGoldProds + MntSweetProducts + MntMeatProducts) as total_revenue 
from 'marketing_data.csv'
order by total_revenue desc 
```

```python
-- Where does our most revenue come from?
select Country
, avg(Income) as avg_income
, sum(MntFishProducts + MntFruits + MntWines + MntGoldProds + MntSweetProducts + MntMeatProducts) as total_revenue 
from 'marketing_data.csv'
group by Country
order by  total_revenue  desc 
```

```python
--What are the factors affecting the number of monthly site visits?
select
    ID,
    case
        when Year_Birth > 1893 and Year_Birth < 1940 then 1
        when Year_Birth >= 1940 and Year_Birth < 1980 then 2
        else 3
    end as classify_of_age,
    Marital_Status,
    Income
from marketing_data.csv
```

```python
-- What is the relationship between income and discounts?
select id 
,Income
,NumDealsPurchases
from 'marketing_data.csv'
order by NumDealsPurchases desc 
```

```python
--How do marketing campaigns affect the response rate?
select ID 
,AcceptedCmp1+AcceptedCmp2+AcceptedCmp3+AcceptedCmp4+AcceptedCmp5 as total_accepts
,Complain
from 'marketing_data.csv'
order by total_accepts desc 
```

```python
--What is the impact of having children and teenagers at home on spending?
select ID
,(NumWebPurchases + NumStorePurchases) as total_purchases
,Kidhome
,Teenhome
from 'marketing_data.csv'
order by total_purchases desc  
```
